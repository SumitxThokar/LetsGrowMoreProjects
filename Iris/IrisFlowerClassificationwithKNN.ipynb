```python
# Importing necessary modules.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score
```

```python
# Header list.
# We know our data don't have a header so we create a list of header and add the header using name parameter.
header_list=['Sepal length','Sepal Width','Petal Length','Petal Width','Species']
```

```python
# Load Data
df=pd.read_csv('iris.csv',names=header_list)
```

## Data Visualization

## Modeling

```python
#X as independent variable and y as dependent variable.
X=df[['Sepal length','Sepal Width','Petal Length','Petal Width']]
y=df[['Species']]
print(X.head())
print(y.head())
```

```python
# Spliting the datasets into train and test sets.
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3)
```

```python
#printing sizes of the sets of data.
print('X_train:',X_train.shape)
print('X_test:',X_test.shape)
print('y_train:',y_train.shape)
print('y_test:',y_test.shape)
```

# Model Selection

```python
KNN=KNeighborsClassifier(n_neighbors=4)
KNN.fit(X_train,y_train)
pred=KNN.predict(X_test)
print('Accuracy of KNN model:',accuracy_score(y_test,pred)*100)
```
